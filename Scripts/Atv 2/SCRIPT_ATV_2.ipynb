```r
# Load data
data <- read.csv("milsa.csv")
milsa <- data

show(data)
```

```r
# Calculate measures of summary and position
summary_stats <- summary(data$Salario)

# Print summary statistics
print(summary_stats)

# Create a histogram to visualize the distribution
hist(data$Salario, main = "Histogram of Salario", xlab = "Salario", col = "lightblue", border = "black")
```

```r
#Measures of Variability
```

```r
# Calculate measures of variability
variance <- var(data$Salario)
std_dev <- sd(data$Salario)

# Print measures of variability
print(paste("Variance:", variance))
print(paste("Standard Deviation:", std_dev))

# Create a box plot to visualize the distribution
boxplot(data$Salario, main = "Box Plot of Salario", xlab = "Salario")
```

```r
# Measures of Form
```

```r
# Calculate measures of form (skewness and kurtosis)
skewness <- skewness(data$Salario)
kurtosis <- kurtosis(data$Salario)

# Print measures of form
print(paste("Skewness:", skewness))
print(paste("Kurtosis:", kurtosis))

# Create a density plot to visualize the distribution
plot(density(data$Salario), main = "Density Plot of Salario", xlab = "Salario")
```

```r
#Qualitative Variable

#Absolute Frequency Distribution
```

```r
# Calculate absolute frequency distribution
freq_dist <- table(data$Inst)

# Print absolute frequency distribution
print(freq_dist)

# Create a bar chart to visualize the distribution
barplot(freq_dist, main = "Absolute Frequency Distribution of Inst", xlab = "Inst", col = "lightblue", border = "black")
```

```r
#Relative Frequency Distribution
```

```r
# Calculate relative frequency distribution
rel_freq_dist <- prop.table(table(data$Inst))

# Print relative frequency distribution
print(rel_freq_dist)

# Create a pie chart to visualize the distribution
pie(rel_freq_dist, main = "Relative Frequency Distribution of Inst", col = rainbow(nlevels(data$Inst)))
```

```r
#Cumulative Frequency Distribution
```

```r
# Calculate cumulative frequency distribution
cum_freq_dist <- cumsum(table(data$Inst))

# Print cumulative frequency distribution
print(cum_freq_dist)

# Create a line chart to visualize the distribution
plot(cum_freq_dist, main = "Cumulative Frequency Distribution of Inst", xlab = "Inst", type = "l")
```

```r
##Medidas de resumo (média, mediana, desvio padrão, mínimo e máximo) da variavel Salario para cada categoria da variável Inst
with(milsa, tapply(Salario, Inst, mean))
with(milsa, tapply(Salario, Inst, median))
with(milsa, tapply(Salario, Inst, sd))
with(milsa, tapply(Salario, Inst, min))
with(milsa, tapply(Salario, Inst, quantile))
with(milsa, tapply(Salario, Inst, max))
```

```r
# Ajustar o modelo de ANOVA para o cálculo do coeficiente de determinação
modelo_anova <- aov(Salario ~ Inst, data = milsa)

anova_sum <- summary(modelo_anova)
ss_total <- sum(anova_sum[[1]][, "Sum Sq"])
ss_modelo <- anova_sum[[1]]["Inst", "Sum Sq"]
r_squared <- ss_modelo / ss_total
r_squared
```
